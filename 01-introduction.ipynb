# Statistical Learning

- Supervised: Predicting output based on inputs
    - Regression: Predict Numerical output
    - Classification: Predict Categorial output
- Unsupervised: Learns relationship from inputs

## History

- Least squares: Eearliest form of Linear Regression
- Linear discriminant analysis in 1936
- Logistic regression in 1940s
- Generalized linear model in 1970s (includes Linear and Logistic regression as special cases)
- Classification and Regression Trees in 1980s (computationally intensive) followed by generalized additive modesl
- Neural networks in 1980s
- Support Vector Machines in 1990s

### Transpose

$
    X^T = \begin{pmatrix}
            x_{11} & x_{22} & \dots & x_{n1} \\
            x_{12} & x_{22} & \dots & x_{n2} \\
            \vdots  & \vdots  & \ddots & \vdots  \\
            x_{1p} & x_{2p} & \dots & x_{np}
        \end{pmatrix}
$
